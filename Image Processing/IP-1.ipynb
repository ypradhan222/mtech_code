```python
import cv2
import matplotlib.pyplot as plt

# specify the image path
im_bgr = cv2.imread('/home/dmacs-5/Desktop/1st Sem/Image Processing/media/download.jpeg')

plt.imshow(im_bgr)
```

```python
# Convert the image from BGR to RGB
im_rgb = cv2.cvtColor(im_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(im_rgb)
plt.axis('off') # turn off the axis
```

```python
# convert the image to grayscale
im_gray = cv2.cvtColor(im_bgr, cv2.COLOR_BGR2GRAY)

plt.imshow(im_gray, cmap = 'gray')
plt.axis('off')
```

```python
# resize an image to (width, height)
im_res = cv2.resize(im_rgb, (300,300))

plt.imshow(im_res)
plt.axis('off')
```

```python
# rotate an image
im_rot = cv2.rotate(im_rgb, cv2.ROTATE_90_COUNTERCLOCKWISE)

plt.imshow(im_rot)
plt.axis('off')
```

```python
# set specific parts of the image to 0
for i in range(20,181):
    for j in range(20,181):
        im_bgr[i][j] = [0,0,0]

plt.imshow(im_bgr)
plt.axis('off')
```

```python
# you can set it to any other color as well
for i in range(20,181):
    for j in range(20,181):
        im_bgr[i][j] = [10,50,67]

plt.imshow(im_bgr)
plt.axis('off')
```

```python
# saving an image
cv2.imwrite('/home/hirak/Desktop/Python/img_bgr.jpg',im_bgr)
```
