```python
import cv2
import matplotlib.pyplot as plt
path = '/home/dmacs-5/Desktop/1st Sem/Image Processing/media/flower2.jpeg'
im = cv2.imread(path)
im_rgb =  cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
im_gray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
plt.imshow(im_rgb)
plt.axis('off')
```

```python
image1 = cv2.imread('/home/dmacs-5/Desktop/1st Sem/Image Processing/media/flower1.jpeg')
image2 = cv2.imread('/home/dmacs-5/Desktop/1st Sem/Image Processing/media/flower2.jpeg')
im_res = cv2.resize(image1,(180,200))
im_res2 = cv2.resize(image2,(180,200))
add_it = cv2.add(im_res,im_res2)
plt.imshow(image1)
plt.title('Image 1')
plt.axis('off')
plt.show()

plt.imshow(image2)
plt.title('Image 2')
plt.axis('off')
plt.show()

plt.imshow(add_it)
plt.title('Addition')
plt.axis('off')
plt.show()
```

```python
imge_sub = cv2.subtract(im_res,im_res2)
plt.imshow(imge_sub)
plt.title('Subtraction')
plt.axis('off')
plt.show()
```

```python
scalar= 0.65
mult =cv2.multiply(im_res,scalar)
plt.imshow(mult)
plt.title('Scalar multiplying')
plt.axis('off')
plt.show()


b = 1
div = cv2.divide(im_res,b)
plt.imshow(div)
plt.title('Division')
plt.axis('off')
plt.show()
```

```python
a= 0.4
b =1-a

addweight = cv2.addWeighted(im_res,a,im_res2,b,0)
plt.imshow(im_res)
plt.axis('off')
plt.title('Image1')
plt.show()

plt.imshow(im_res2)
plt.axis('off')
plt.title('Image2')
plt.show()

plt.imshow(addweight)
plt.axis('off')
plt.title('Weighted Image')
plt.show()

```
